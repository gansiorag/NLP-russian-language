```python
def allTegPath(IsTxt:str, nP:str):
    tegs = []
    tt = IsTxt.split(f'<{nP}>')[1].split(f'</{nP}>')[0]
    ttl = tt.split('<')
    for kk in ttl:
        if '>' in kk and '/' not in kk:
            #if '/' in kk and 'http' not in kk:
                iti = kk.split('>')[0].strip().replace('>','')
                if ' ' in iti:
                    iti = iti.split(' ')[0]
                tegs.append(iti)
    return tegs
```

```python
def author(IsTxt:str) -> dict:
    DataDict={'first-name':'',
             'middle-name':'',
             'last-name':'',
             'nickname':'',
             'home-page':'', 
             'id>':''
             }
    tt = IsTxt.split('<author>')[1].split('</author>')[0]
#     print(tt)
#     print()
    for kk in DataDict:
        if kk in tt:
            if kk != 'home-page':
                DataDict[kk] = tt.split(kk)[1].replace('<','').replace('>','').replace('/','')
            else:
                DataDict[kk] = tt.split(kk)[1].replace('<','').replace('>','')[:-1]
    print(DataDict)
    return DataDict
```

```python
def getTXT(ComText:str):
    tokSentens = []
    dirtyTxt = ComText.split('<section>')[1].split('</section>')[0]
    dirtyTxtList = dirtyTxt.split('<p>')
    for ss in dirtyTxtList:
        Clss = ss.replace('</p>', '').replace('\n', '').replace('<STRONG>', ' ').replace('</STRONG>', ' ').strip()
        Clss = re.sub(r'[,.!–?><«»";:…-]', ' ', Clss)
        Clss = Clss.upper().replace('STRONG', ' ').replace('/STRONG', ' ').replace('LINE', ' ').replace('EMPTY', ' ')
        Clss = Clss.replace('/', ' ')
        Clss = re.sub(r'\s+', ' ', Clss)
        if len(Clss)>0:
            tokSentens.append(Clss.strip())
    return tokSentens    
```

```python
wokab = Counter()
for nf in nFList:
    with open(nf, 'r') as i_f:
        comText = i_f.read()
    txtList = getTXT(comText)
    for senten in txtList:
        arrWord = senten.split()
        for word in arrWord:
            wokab[word] += 1
#     print(len(txtList))
#     print(txtList)
    #break
```
